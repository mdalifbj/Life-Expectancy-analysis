```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score
from sklearn.metrics import confusion_matrix, ConfusionMatrixDisplay
```

```python
data_last = data_last.join(data[remove_list])
data_last.head()
```

```python
X = data_last.drop('Life expectancy', axis=1)
y = data_last['Life expectancy']
```

```python
X['Year'] = X['Year'].astype('str')

X = pd.get_dummies(X, drop_first=True)
X.head()
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
print(X_scaled)
```

```python
X_train,X_test,y_train,y_test = train_test_split(X_scaled,y,test_size=0.2,random_state=42)
ml = LinearRegression()
ml.fit(X_train,y_train)
```

```python
print('Coefficient of determination: ', ml.score(X_train, y_train))
```

```python
y_pred = ml.predict(X_train)
```

```python
print('R^2: ', r2_score(y_train, y_pred))
print('MAE: ',mean_absolute_error(y_train, y_pred))
print('MSE: ', mean_squared_error(y_train, y_pred))
print('RMSE: ', np.sqrt(mean_squared_error(y_train, y_pred)))
```

```python
print('Coefficient of determination: ', ml.score(X_test, y_test))
```

```python
y_pred_test = ml.predict(X_test)
```

```python
print('R^2: ', r2_score(y_test, y_pred_test))
print('MAE: ',mean_absolute_error(y_test, y_pred_test))
print('MSE: ', mean_squared_error(y_test, y_pred_test))
print('RMSE: ', np.sqrt(mean_squared_error(y_test, y_pred_test)))
```

```python
results = pd.DataFrame({'actual': y_test, 
              'predicted': y_pred_test.ravel(), 
              'residual': y_test - y_pred_test})
results.head(22)
```

```python
sb.scatterplot(x=results['actual'], 
                y=results['predicted'])
plt.title('Actual vs Predicted values')
plt.show()
```

```python
sb.histplot(results['residual'], bins=np.arange(-10,10,0.5))
plt.title('Histogram of residuals')
plt.show()
```
